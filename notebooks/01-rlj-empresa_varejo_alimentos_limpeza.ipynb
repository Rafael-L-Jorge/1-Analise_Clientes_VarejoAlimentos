Ao verificar os dados nulos e o conteúdo das suas colunas, decidi transformar os dados em booleanos para melhorar a análise.
O conteúdo das colunas era basicamente composto por: o código que representava o valor verdadeira da coluna e NaN.
Exemplo: na coluna education_graduation, tinhamos o valor 3.0 quando o cliente tinha graduation e NaN quando não tinha. 

```python
#Colunas com nulos
colunas_com_nulos = ["marital_Divorced", 
                     "marital_Married", 
                     "marital_Single", 
                     "marital_Together", 
                     "marital_Widow", 
                     "education_2n Cycle", 
                     "education_Basic", 
                     "education_Graduation", 
                     "education_Master", 
                     "education_PhD"]

#substituindo por booleano
for item in colunas_com_nulos:
  df["booleano."+str(item)] = np.where(df[item].isnull(), 0, 1) #se for nulo, substituimos por 0, caso contrário substituimos por 1
```

```python
with pd.option_context("display.max_columns", None):
    display(df.head())
```

```python
#Exclusão de coluna substituidas por booleanos
df = df.drop(
    columns=["marital_Divorced", 
             "marital_Married", 
             "marital_Single", 
             "marital_Together", 
             "marital_Widow", 
             "education_2n Cycle", 
             "education_Basic", 
             "education_Graduation", 
             "education_Master", 
             "education_PhD"]
)

with pd.option_context("display.max_columns", None):
    display(df.head())
```

Julguei necessária a tradução de duas colunas especificadas, sendo elas: estado civil e nivel de escolaridade.

```python
# traduzir os valores de estado cívil
estadocivil_traducao = {
    'Married':   'Casado(a)',
    'Together':  'Uniao estavel',
    'Single':    'Solteiro(a)',
    'Divorced':  'Divorciado(a)',
    'Widow':     'Viuvo(a)'
}

#renomeando os valores da coluna estado civil
df['estado_civil(t)'] = (
    df['estado_civil']          # coluna original
            .replace(estadocivil_traducao)   # aplicar tradução
)

with pd.option_context("display.max_columns", None):
    display(df.head())
```

```python
# traduzir os valores de escolaridade
escolaridade_traducao = {
    'Graduation': 'Graduacao',
    'PhD': 'Doutorado',
    'Master': 'Mestrado',
    '2n Cycle': '2n ciclo',
    'Basic': 'Básica'
}

#renomeando os valores da coluna escolaridade
df['nivel_escolaridade(t)'] = (
    df['nivel_escolaridade']          # coluna original
            .replace(escolaridade_traducao)   # aplicar tradução
)

with pd.option_context("display.max_columns", None):
    display(df.head())
```

Com a tradução dessas colunas, vou excluir as originais.

```python
#Exclusão das colunas de escolaridade e estado civil
df = df.drop(
    columns=[
        'nivel_escolaridade', 
        'estado_civil']
)

with pd.option_context("display.max_columns", None):
    display(df.head())
```
